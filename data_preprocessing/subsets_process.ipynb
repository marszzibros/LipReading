```python
import pandas as pd 
```

```python
df = pd.read_csv("metadata.csv", index_col=0)
```

```python
df
```

```python
df['word'] = df['out_dir'].apply(lambda x: x.split('/')[-1].split('_')[1])
```

```python
fifty_words = df['word'].value_counts().keys()[0:50]
```

```python
new_df = df[df['word'].apply(lambda x: x in fifty_words)]
```

```python
new_df
```

```python
new_df['type'] = new_df['out_dir'].apply(lambda x: x.split('/')[-1].split('_')[0])
```

```python
new_df
```

```python
new_df.groupby(['word', 'type']).agg({'word':'count'})
```

```python
new_df.reindex()
```

```python
new_df.to_csv("subset_test.csv")
```
