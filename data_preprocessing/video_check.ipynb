<h2> Basic Information about a video </h2>

*   The number of Frames and Durations is different in each video as length of the sentence is different.
*   At least in Google Colab, it takes almost 6 seconds to process and display one video. As a result, we may need to pre-process (i.e. by frame) them before training.


```python
start = time.time()
cap = cv2.VideoCapture(os.path.join(GOOGLE_DRIVE_PATH, 'sample_data','LRS2', sample_video))

# Get the video details
if cap.isOpened():
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count / frame_rate

    print(f"Width: {width}")
    print(f"Height: {height}")
    print(f"Frame Rate: {frame_rate}")
    print(f"Frame Count: {frame_count}")
    print(f"Duration (seconds): {duration}")
else:
    print("Error opening video file")

cap.release()
end = time.time()

print(f"Time: {end - start}")
```

```python
frames_to_display = 40
count = 0

start = time.time()
fig, axes = plt.subplots(8, 5, figsize=(15, 24))
cap = cv2.VideoCapture(os.path.join(GOOGLE_DRIVE_PATH, 'sample_data','LRS2', sample_video))

for i in range(8):
    for j in range(5):
        ret, frame = cap.read()

        if not ret:
            break

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        axes[i, j].imshow(frame_rgb)
        axes[i, j].axis('off')

        count += 1
        if count >= frames_to_display:
            break

cap.release()

plt.tight_layout()
plt.show()

end = time.time()

print(f"Time: {end - start}")
```

<h2>Different Vidoes, Different Specs</h2>

```python

for file_name in os.listdir(os.path.join(GOOGLE_DRIVE_PATH, 'sample_data', 'LRS2')):
  if file_name.endswith('.mp4'):

    cap = cv2.VideoCapture(os.path.join(GOOGLE_DRIVE_PATH, 'sample_data','LRS2', file_name))

    # Get the video details
    if cap.isOpened():
        frame_rate = cap.get(cv2.CAP_PROP_FPS)
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        duration = frame_count / frame_rate

        print(f"file: {file_name}")
        print(f"Frame Rate: {frame_rate}")
        print(f"Frame Count: {frame_count}")
        print(f"Duration (seconds): {duration}")
    else:
        print("Error opening video file")

    cap.release()
```
